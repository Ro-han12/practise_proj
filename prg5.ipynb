```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("adult.csv")
data.head()
```

```python
data['workclass']=data['workclass'].replace('?',np.nan)
data['occupation']=data['occupation'].replace('?',np.nan)
data['native-country']=data['native-country'].replace('?',np.nan)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data=data.drop_duplicates()
```

```python
data.dropna(how='any',axis=1,inplace=True)
```

```python
data.head()
```

```python
data.drop(['fnlwgt','educational-num','marital-status','relationship','race'],axis=1,inplace=True)
```

```python
data.head()
```

```python
x=data.drop(columns='income')
y=data['income']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y=le.fit_transform(y)
y=pd.DataFrame(y)
y.head()
```

```python
numeric_features=x.select_dtypes('number')
categorical_features=x.select_dtypes('object')
```

```python
categorical_features
```

```python
categorical_features.shape
```

```python
converted_cf=pd.get_dummies(categorical_features)
converted_cf.head()
```

```python
all=[converted_cf,numeric_features]
new=pd.concat(all,axis=1,join='inner')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(new,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier as DTC
clf=DTC(max_depth=5) #max_depth=5 indicates from root node to leaf node we can have 5 nodes
clf.fit(x_train,y_train)
```

```python
#predicting test results
y_pred = clf.predict(x_test)
print(y_test.shape)
print(y_pred.shape)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plot_tree(clf,feature_names=None,class_names=['0','1'],rounded=True,filled=True)
plt.show()
```
